```python
import Utils.superResolution as sr
import cv2
import matplotlib.pyplot as plt
import plotly.express as px

import numpy as np

```

```python
image_low_res = cv2.imread('images/lowRes.jpg')

lowResImg = image_low_res[90:120 , 160:240]
```

```python
px.imshow(image_low_res)

# plt.subplot(122)
# plt.imshow(lowResImg)
```

```python
fig = plt.figure(figsize=(100, 100), dpi=100)

fig.add_subplot(1,2, 1)
plt.imshow(image_low_res, cmap='gray')
plt.title('image_low_res')

fig.add_subplot(1,2, 2)
plt.imshow(lowResImg, cmap='gray')
plt.title('lowResImg')
```

```python
cubic = sr.superResCUBIC(lowResImg)
area = sr.superResAREA(lowResImg)
BITS = sr.superResBITS(lowResImg)
BITS2 = sr.superResBITS2(lowResImg)
LANCZOS4 = sr.superResLANCZOS4(lowResImg)
LINEAR = sr.superResLINEAR(lowResImg)
LINEAR_EXACT = sr.superResLINEAR_EXACT(lowResImg)
NEAREST = sr.superResNEAREST(lowResImg)
NEAREST_EXACT = sr.superResNEAREST_EXACT(lowResImg)
```

```python
fig = plt.figure(figsize=(100, 100), dpi=100)

fig.add_subplot(4, 3, 1)
plt.imshow(lowResImg, cmap='gray')
plt.title('lowResImg')

fig.add_subplot(4, 3, 2)
plt.imshow(cubic, cmap='gray')
plt.title('cubic')

fig.add_subplot(4, 3, 3)
plt.imshow(area, cmap='gray')
plt.title('area')

fig.add_subplot(4, 3, 4)
plt.imshow(BITS, cmap='gray')
plt.title('BITS')

fig.add_subplot(4, 3, 5)
plt.imshow(BITS2, cmap='gray')
plt.title('BITS2')

fig.add_subplot(4, 3, 6)
plt.imshow(LANCZOS4, cmap='gray')
plt.title('LANCZOS4')

fig.add_subplot(4, 3, 7)
plt.imshow(LINEAR, cmap='gray')
plt.title('LINEAR')

fig.add_subplot(4, 3, 8)
plt.imshow(LINEAR_EXACT, cmap='gray')
plt.title('LINEAR_EXACT')

fig.add_subplot(4, 3, 9)
plt.imshow(NEAREST, cmap='gray')
plt.title('NEAREST')

fig.add_subplot(4, 3, 10)
plt.imshow(NEAREST_EXACT, cmap='gray')
plt.title('NEAREST_EXACT')


```
